### Problem: The Movie Revenue Dataset

**Dataset Description**

This dataset has been obtained from https://www.the-numbers.com/movie/budgets

This data has already been saved in a csv file called 'Dataset_Movie_Revenue.csv' and provided with this notebook.

The csv file contains 6 columns (A to F) as follows:
- Column A: Rank or serial number of the movie as per the production budget. Movies are ranked in decreasing order of their respective production budgets.
- Column B: Release date of the movie.
- Column C: Title of the movie.
- Column D: Production budget of the movie in dollars.
- Column E: Domestic gross in dollars made by the movie.
- Column F: Worldwide gross in dollars made by the movie.

Let us try all these cleaning steps using **Python**.

Save the file as 'Dataset_Movie_Revenue_Clean_Python.csv'.

```python
#Remove all other irrelevent columns, i.e. remove coulumns A, B, C, and E.
data=data.drop(['Rank','Release Date', 'Title', 'Domestic Gross'], axis=1)
data
```

```python
#Change the titles of the remaining two columns to 'Production_Budget' and 'World_Gross' (remove spaces from names).
data = data.rename(columns={'Production Budget':'Production_Budget', 'Worldwide Gross': 'World_Gross'})
data
```

```python
#Since both the columns should contain numbers, remove the commas from the values.
data['Production_Budget'] = (data['Production_Budget'].astype(str).str.replace(',', ''))
data['Production_Budget'] = pd.to_numeric(data['Production_Budget'], errors='coerce')
data['World_Gross'] = (data['World_Gross'].astype(str).str.replace(',', ''))
data['World_Gross'] = pd.to_numeric(data['World_Gross'], errors='coerce')
print(data.info())
data
```

```python
#Delete all rows where the data is missing, i.e. delete all the rows where 'World_Gross' is 0.
data = data.drop(data[data['World_Gross'] == 0].index)
data
```

```python
#How many rows did u delete? Is this number significant as compared to the total dataset size?
print(6169-len(data))
# Since this is a very small percentage of teh total dataset, it is not significant.
```

```python
#All the changes that we have made so far are done on the dataframe, and not in the original csv file.
#The to_csv method can be used to save the dataframe into a csv file.
#Lets save this clean data into a new file 'D1_Movie_Revenue_Dataset_Clean_Python.csv'
data.to_csv('Dataset_Movie_Revenue_Clean_Python.csv', index=False)
```

### Task for you

Repeat the cleaning process in Python,
this time instead of deleting the '0' columns, replace the zeros with average value of the column.

Tryout Pandas DataFrame’s fillna() method.


```python
import pandas as pd
import numpy as np

# Re-reading the original file to start fresh for the new cleaning process
data_new_clean = pd.read_csv('Dataset_Movie_Revenue.csv')
display(data_new_clean.head())
```

```python
# Remove all other irrelevant columns
data_new_clean = data_new_clean.drop(['Rank','Release Date', 'Title', 'Domestic Gross'], axis=1)

# Change the titles of the remaining two columns
data_new_clean = data_new_clean.rename(columns={'Production Budget':'Production_Budget', 'Worldwide Gross': 'World_Gross'})
display(data_new_clean.head())
```

```python
# Remove commas from the values and convert to numeric
data_new_clean['Production_Budget'] = (data_new_clean['Production_Budget'].astype(str).str.replace(',', ''))
data_new_clean['Production_Budget'] = pd.to_numeric(data_new_clean['Production_Budget'], errors='coerce')

data_new_clean['World_Gross'] = (data_new_clean['World_Gross'].astype(str).str.replace(',', ''))
data_new_clean['World_Gross'] = pd.to_numeric(data_new_clean['World_Gross'], errors='coerce')

print("DataFrame Info after numeric conversion:")
data_new_clean.info()
display(data_new_clean.head())
```

```python
# Replace '0' values in 'World_Gross' with NaN first
data_new_clean['World_Gross'] = data_new_clean['World_Gross'].replace(0, np.nan)

# Fill the NaN values with the calculated mean
data_new_clean['World_Gross'] = data_new_clean['World_Gross'].fillna(data_new_clean['World_Gross'].mean())

print("\nDataFrame Info after filling 0s with mean:")
data_new_clean.info()
display(data_new_clean)
```

```python
# Save this cleaned data into a new file 'Dataset_Movie_Revenue_Clean_Python_Mean_Fill.csv'
data_new_clean.to_csv('Dataset_Movie_Revenue_Clean_Python_Mean_Fill.csv', index=False)

print("Cleaned data saved to 'Dataset_Movie_Revenue_Clean_Python_Mean_Fill.csv'")
```

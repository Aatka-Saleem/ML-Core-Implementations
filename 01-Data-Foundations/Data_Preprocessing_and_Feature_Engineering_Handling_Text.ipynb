## TEXT PREPROCESSING AND AUGMENTATION

### The Dataset

- IMBd (Internet Movie Database) is an online database of information related to films, television series, podcasts, home videos, video games, and streaming content online, including cast, production crew and personal biographies, plot summaries, trivia, ratings, and fan and critical reviews.
- Keras has a built-in IMDb movie reviews data set.
- This is a dataset of 50,000 movies reviews from IMDB, labeled by sentiment (positive/negative).
- They are split into 25,000 for training and 25,000 for testing, each set consisting of 50% negative and 50% positive reviews.
- Reviews have been tokenized, and each review is encoded as a list of word indexes (integers).
- For convenience, words are indexed by overall frequency in the dataset, so that for instance the integer "3" encodes the 3rd most frequent word in the data. This allows for quick filtering operations such as: "only consider the top 10,000 most common words, but eliminate the top 20 most common words".
- Indices are off by 3 because 0, 1, and 2 are reserverd indices for "padding", "Start of sequence" and "unknown" / out-of-vocabulary (oov).

**Loading the dataset and setting vocabulary size**

```python
from keras.datasets import imdb

vocabulary_size = 5000

(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words = vocabulary_size)
# num_words: integer or None.
# Words are ranked by how often they occur (in the training set) and
# Only the num_words most frequent words are kept.
# Any less frequent word will appear as oov_char (out of vocabulary) value in the sequence data.
# If None, all words are kept. Defaults to None.

print(f'Loaded dataset with {len(X_train)} training samples, {len(X_test)} test samples')
```

**Checking out a Review**

```python
# Checking out a review
review_no=6
print('Review:')
print(X_train[review_no])
print('Label:')
print(y_train[review_no])
```

- Note that the review is stored as a sequence of integers.
- These are word IDs that have been pre-assigned to individual words, and the label is an integer (0 for negative, 1 for positive).
- We can use the dictionary returned by imdb.get_word_index( ) to map the review back to the original words.

```python
# Mapping the review back to the original words

# Loading the dictionary mappings from word to integer index
word_index = imdb.get_word_index()

# Reversing word index to map integer indexes to their respective words
reverse_word_index = dict([(value, key) for (key, value) in word_index.items()])

# Decoding the review, mapping integer indices to words
decoded_review = ' '.join([reverse_word_index.get(i-3, '?') for i in X_train[review_no]])

decoded_review
```

**Padding Sequences**
- In order to feed this data into a neural network, all input documents must have the same length.
- We will limit the maximum review length to max_words by truncating longer reviews and padding shorter reviews with a null value (0).
- We can accomplish this using the pad_sequences( ) function in Keras.
- For now, set max_words to 500.

```python
# Finding maximum review length and minimum review length
print('Maximum review length: {}'.format(len(max((X_train + X_test), key=len))))
print('Minimum review length: {}'.format(len(min((X_test + X_test), key=len))))
```

```python
from keras.utils import pad_sequences

max_words = 500
X_train = pad_sequences(X_train, maxlen=max_words)
X_test = pad_sequences(X_test, maxlen=max_words)
```
